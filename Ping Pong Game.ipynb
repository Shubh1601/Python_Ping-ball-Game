```python
import pygame
import sys

# Initialize pygame
pygame.init()

# Set up display
WIDTH, HEIGHT = 800, 600
win = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Ball Bounce Game")

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)

# Paddle
paddle_width, paddle_height = 100, 10
paddle_x = (WIDTH - paddle_width) // 2
paddle_y = HEIGHT - 40
paddle_speed = 7

# Ball
ball_radius = 10
ball_x = WIDTH // 2
ball_y = HEIGHT // 2
ball_dx = 4
ball_dy = -4

# Clock
clock = pygame.time.Clock()

# Score
score = 0
font = pygame.font.SysFont("comicsans", 30)

# Main Game Loop
running = True
while running:
    clock.tick(60)
    win.fill(BLACK)

    # Event Handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Key Handling
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and paddle_x > 0:
        paddle_x -= paddle_speed
    if keys[pygame.K_RIGHT] and paddle_x < WIDTH - paddle_width:
        paddle_x += paddle_speed

    # Move Ball
    ball_x += ball_dx
    ball_y += ball_dy

    # Collision with walls
    if ball_x - ball_radius <= 0 or ball_x + ball_radius >= WIDTH:
        ball_dx *= -1
    if ball_y - ball_radius <= 0:
        ball_dy *= -1

    # Collision with paddle
    if (paddle_y <= ball_y + ball_radius <= paddle_y + paddle_height) and (paddle_x <= ball_x <= paddle_x + paddle_width):
        ball_dy *= -1
        score += 1

    # Ball missed
    if ball_y > HEIGHT:
        running = False

    # Draw paddle
    pygame.draw.rect(win, WHITE, (paddle_x, paddle_y, paddle_width, paddle_height))

    # Draw ball
    pygame.draw.circle(win, WHITE, (ball_x, ball_y), ball_radius)

    # Draw score
    score_text = font.render(f"Score: {score}", True, WHITE)
    win.blit(score_text, (10, 10))

    # Update display
    pygame.display.update()

# Game Over
win.fill(BLACK)
game_over_text = font.render("Game Over! Press any key to exit.", True, WHITE)
win.blit(game_over_text, (WIDTH // 2 - game_over_text.get_width() // 2, HEIGHT // 2))
pygame.display.update()

# Wait for key press to exit
waiting = True
while waiting:
    for event in pygame.event.get():
        if event.type == pygame.KEYDOWN or event.type == pygame.QUIT:
            waiting = False

pygame.quit()
sys.exit()
```
